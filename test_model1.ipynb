```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
```

```python
df = pd.read_csv("GN_01_06_23.csv", header=0, names=None)
df.head()
```

```python
columns_to_drop = ['ID', 'Isolate Number','HN','Unit','Methods','Vitek card lot','Sex',]

df = df.drop(columns=columns_to_drop)
df.columns
```

```python
df = pd.get_dummies(df, columns=['Species','Submitted sample', 'Bacterial species identification'])
```

```python
df.head()
```

```python
columns = df.columns
columns_med = []

for i in range(0,len(columns)):
    column_name=columns[i]
    column_name.strip()
    if column_name[-1]=="_":
        columns_med.append(column_name)
    
print(columns_med)


```

```python
featurescol = []

for col in df.columns:
    if "Species" in col or "Submitted sample" in col or "Bacterial species identification" in col:
        featurescol.append(col)

```

```python
model_name ="Amikacin_"
path="../2021_Dataset/models/model_"+model_name+"gn.pkl"

model=joblib.load(path)
```

```python
for col in columns_med:
    if col == model_name.lower():
        featurescol.append(col)

print(featurescol)
```

```python
# Perform one-hot encoding
df=df[featurescol]
one_hot_encoded = pd.get_dummies(df['amikacin_'], prefix='S/I/R')

# Concatenate the one-hot encoded columns to the original DataFrame
df = pd.concat([df, one_hot_encoded], axis=1)
df.head()
```
