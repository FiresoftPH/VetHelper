```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import joblib
```

```python
df = pd.read_csv("gn_test.csv", header=0, names=None)
df.head()
```

```python
columns_to_drop = ['Breed','ID', 'Isolate Number','HN','Unit','Methods','Vitek card lot','Sex','Method of collection','Counting (CFU/mL)']

df = df.drop(columns=columns_to_drop)
df.columns
```

```python
df.rename(columns={'Species': "species"}, inplace=True)
df.rename(columns={'Submitted sample':  "submitted_sample"}, inplace=True)
df.rename(columns={'Bacterial species identification': "bacteria_genus"}, inplace=True)
df['species'] = df['species'].str.lower()
```

```python
columns = df.columns
columns_med = []

for i in range(0,len(columns)):
    column_name=columns[i]
    column_name.strip()
    if column_name[-1]=="_":
        columns_med.append(column_name)
    
print(columns_med)


```

```python
featurescol = []
for col in df.columns:
    if "species" in col or "submitted_sample" in col or "bacteria_genus" in col:
        featurescol.append(col)
print(featurescol)
```

```python
for col in columns_med:

    featurescol.append(col)

print(featurescol)
```

```python
# Perform one-hot encoding
df=df[featurescol]
df.head()
for column in df.columns:
    df[column] = df[column].astype(str).replace("*R", "R")
    if column[-1]=="_":
        df[column] = df[column].str.replace('_', '')
        #df[column] = df[column].astype(str).replace("_", "")
    
df.columns
```

```python
for index, row in df.iterrows():
    # Iterate through each element in the row
    for column, value in row.items():
        
        if value == "NA":
            df = df.drop(index)

   
```

```python
for i in range(0,len(columns_med)):
    med_name = columns_med[i]
    med_name = med_name.replace("_", "")
    df = pd.get_dummies(df,columns=[columns_med[i]],prefix="S/I/R_"+med_name)
# Concatenate the one-hot encoded columns to the original DataFrame

df.head()
df.columns
```
