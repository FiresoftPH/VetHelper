```python
from pdfminer.high_level import extract_pages,extract_text


text = extract_text("pdf_test_file.pdf")
print(text)
```

```python
import re

species_pattern =  re.compile(r'(\w+)\s+(?=(Male|Female))', re.MULTILINE)
matches = species_pattern.findall(text)
species=matches[0][0]
gender=matches[0][1]

print(species)
print(gender)
```

```python
import re

text = """Cat
Male

Patient
Breed
Birth date"""

pattern = re.compile(r'(\w+)\s+(?=(Male|Female))', re.MULTILINE)

matches = pattern.search(text)

if matches:
    word_before_gender = matches.group(1)
    print(word_before_gender)
```
