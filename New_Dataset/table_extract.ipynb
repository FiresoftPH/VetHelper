```python
from pdfminer.high_level import extract_pages,extract_text


text = extract_text("pdf_test_file2.pdf")

words = text.split('\n')
for i in range(len(words)):
    words[i]=words[i].rstrip()
print(words)
```

```python
first_med_index_left_table = words.index('ESBL')
left_table_med = []
for i in range(first_med_index_left_table,first_med_index_left_table+10):
    left_table_med.append(words[i])

print(left_table_med)
```

```python
first_med_index_right_table = words.index('Gentamicin')
right_table_med = []
for i in range(first_med_index_right_table,first_med_index_right_table+10):
    right_table_med.append(words[i])

print(right_table_med)
```

```python
SIR_value_found = False
#assign number 2 to s i r value in the table on the right side of pdf
for i in range(len(words)):
    if words[i] == 'S/I/R' and SIR_value_found == False:
        SIR_value_found = True
    elif words[i] == 'S/I/R' and SIR_value_found == True:
        words[i] = 'S/I/R2'
```

```python
print(words)
```

```python
left_sir = words.index('S/I/R')
left_sir_arr = []
for i in range(left_sir+1,left_sir+11):
    left_sir_arr.append(words[i])
print(left_sir_arr)
```

```python
right_sir = words.index('S/I/R2')
right_sir_arr = []
for i in range(right_sir+1,right_sir+11):
    right_sir_arr.append(words[i])
print(right_sir_arr)
```

```python
result_dict = {left_med: left_sir for left_med, left_sir in zip(left_table_med, left_sir_arr)}

# Print the resulting dictionary
print(result_dict)
```

```python
result_dict = {right_med: right_sir for right_med, right_sir in zip(right_table_med, right_sir_arr)}

# Print the resulting dictionary
print(result_dict)
```
