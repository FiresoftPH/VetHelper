```python
import pandas as pd
import numpy as np

# Load the original dataset
df = pd.read_excel("GN_Dataset_1.xlsx")
# Load the new dataset
new_df = pd.read_csv("GN_01_06_23.csv")
```

```python
new_df.columns = new_df.columns.str.lower()
# Change the bactierial species identification column name to bacteria_genus
new_df = new_df.rename(columns={"bacterial species identification": "bacteria_genus"})
# Change the submitted sample column name to sumbitted_sample
new_df = new_df.rename(columns={"submitted sample": "submitted_sample"})
new_df.columns
```

Edit the two blocks below to suit the GP dataset (change the values or match the column names etc)

```python
# Change the column names to match the original dataset
# From 'cefalothin' to 'cefalotin'
new_df = new_df.rename(columns={"cefalothin": "cefalotin"})
# From 'cephalexin'to 'cefalexin'
new_df = new_df.rename(columns={"cephalexin": "cefalexin"})
# 'sulfamethoxazole/trimethoprim' to 'trimethoprim/sulfamethoxazole'
new_df = new_df.rename(columns={"sulfamethoxazole/trimethoprim": "trimethoprim/sulfamethoxazole"})
```

```python
# Modify the data in the esbl column from the value POS to + and NEG to -
new_df["esbl"] = new_df["esbl"].str.replace("POS", "+").str.replace("NEG", "-")
new_df["esbl"].value_counts(), new_df["esbl"].unique()
```

```python
# Find the columns that has "_" in the column name and append S/I/R_ to the column name and drop the _ from the column name
for col in new_df.columns:
    if "_" in col and col != "bacteria_genus" and col != "submitted_sample":
        new_df = new_df.rename(columns={col: f"S/I/R_{col.replace('_', '')}"})

new_df.columns
```

```python
# Select the features column from the new dataset and the old dataset (For the new dataset, also includes the target for testing)
features_col = ["interpretation and recommendation"]
for col in new_df.columns:
    if "species" in col or "submitted_sample" in col or "bacteria_genus" in col or "S/I/R_" in col:
        features_col.append(col)

new_df = new_df[features_col]

features_col = []
for col in df.columns:
    if "species" in col or "submitted_sample" in col or "bacteria_genus" in col or "S/I/R_" in col:
        features_col.append(col)

df = df[features_col]
new_df.columns, df.columns
```

```python
# Change the uppercase values in the species column to lowercase
new_df["species"] = new_df["species"].str.lower()
# Change the value in the submitted_sample column to lowercase
new_df["submitted_sample"] = new_df["submitted_sample"].str.lower()
# Change the value of the bacteria_genus column to only include the genus name (which was the first word in the original value)
new_df["bacteria_genus"] = new_df["bacteria_genus"].str.split(" ").str[0]
new_df["bacteria_genus"] = new_df["bacteria_genus"].str.lower()
# Look at the unique values in the bacteria_genus column
new_df["bacteria_genus"].unique()
```

```python
# Save the target column from the new dataset and drop it from the main dataframe
target = new_df["interpretation and recommendation"]
new_df = new_df.drop(columns=["interpretation and recommendation"])
```

```python
# Perform one-hot encoding on the original dataset and the new dataset
df = pd.get_dummies(df, columns=["species", "submitted_sample", "bacteria_genus"])
for column in df.columns:
    if "S/I/R_" in str(column):
        df = pd.get_dummies(df, columns=[column])

new_df = pd.get_dummies(new_df, columns=["species", "submitted_sample", "bacteria_genus"])
for column in new_df.columns:
    if "S/I/R" in str(column):
        new_df = pd.get_dummies(new_df, columns=[column])
```

```python
# Compare the new dataset columns with the old dataset columns.
# If the the new dataset columns are not in the old dataset columns, remove them from the dataframe of the new dataset.
# If the the old dataset columns are not in the new dataset columns, add them to the dataframe of the new dataset with the value of False
# since the new dataset does not have the column and the old dataset has the column
for column in df.columns:
    if column not in new_df.columns:
        new_df[column] = False
        print(new_df[column].sample(5))

for column in new_df.columns:
    if column not in df.columns:
        new_df = new_df.drop(columns=[column])
```

```python
# Compare the new dataset columns with the old dataset columns.
new_df.columns.shape, df.columns.shape
```

```python
"""
This creates the sample dataset to be used for testing the model and writes it to a file. (Uncomment the code to run it)
1. Create a file named new_samples.xlsx (Optional: If the file already exists, it will be overwritten)
2. Create another file named model_evaluation_results + bacteria_gram + .xlsx (Mandoratory)
"""
bacteria_gram = "gn"
# Combine the new dataset features and its target
new_df = pd.concat([new_df, target], axis=1)
# Take 20 samples from the new dataset and predict the values using the saved model
new_df = new_df.sample(20)
# Save the samples to a csv file
new_df.to_excel("new_samples.xlsx", index=False)
# Write the target values to a excel file
results_file_name = "model_evaluation_results_" + bacteria_gram + ".xlsx"
target = pd.read_excel("new_samples.xlsx")
results_csv = pd.read_excel(results_file_name)
results_csv["interpretation and recommendation"] = target["interpretation and recommendation"]
results_csv.to_excel(results_file_name, index=False)
```

```python
# Load the saved model
import joblib
directory = "../2021_Dataset/models"
medicine_array = allmeds = ["Trimethoprim-sulfamethoxazole","Cefovecin","Marbofloxacin","Amikacin","Amoxicillin-clavulanic acid","Imipenem","Enrofloxacin","Cefalexin","Gentamicin","Nitrofurantoin","Doxycycline"]
for medicine_name in medicine_array:
    model_filename = "model" + "_" + medicine_name + "_" + bacteria_gram + ".pkl"
    model = joblib.load(f"{directory}/{model_filename}")
    # Load the new samples
    new_df = pd.read_excel("new_samples.xlsx")
    # Select the features from the new samples
    new_df_X = new_df.drop(columns=["interpretation and recommendation"])
    # Select the target from the new samples
    new_df_y = new_df["interpretation and recommendation"]
    # Sort the models feature names and the new dataset columns to match
    model.feature_names_in_ = np.sort(model.feature_names_in_)
    new_df_X = new_df_X[model.feature_names_in_]
    # Predict the values
    predictions = model.predict(new_df_X)
    print(predictions)

    # Write the predictions to a file
    results_file_name = "model_evaluation_results_" + bacteria_gram + ".xlsx"
    results_csv = pd.read_excel(results_file_name)
    results_csv[medicine_name] = predictions
    results_csv.to_excel(results_file_name, index=False)
```
